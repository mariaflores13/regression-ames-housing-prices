## Modeling

```python
import pandas as pd  
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle

from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score

sns.set_style(style = 'whitegrid')
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
clean_test = pd.read_csv('../datasets/clean_test.csv')
```

```python
train = pickle.load(open("../pickles/train_clean.pkl", "rb"))
```

```python
train.isnull().sum().sum()
```

Create `saleprice_log` column in training dataset.
- I was curious to see how taking the log of `saleprice` would affect the model and it improved my RMSE score by $10,000.

```python
# add new 'saleprice_log' column
train['saleprice_log'] = np.log(train['saleprice'])
```

**Features** I have chosen for my model. I will also add `neighborhood_cols` list to the features list.
- Despite the neeighborhood dummies not having as big of an effect on feature correlation with sale price, I still suspect location is important and should be in cluded in my model.

```python
features = ['overall_qual', 'year_built', 'year_remod/add', 'exter_qual', 'bsmt_qual', 'bed_bath_ratio',
            'bsmt_cond', 'bsmtfin_sf_1', 'full_bath', 'kitchen_qual', 'totrms_abvgrd', 'garage_cars', 
            'garage_area', 'garage_qual_cars', 'overall_cond_bed_bath', 'gr_liv_area', 'overall_cond']
# Define list of neighborhood dummies
neighborhood_cols = [col for col in train.columns if 'neighborhood' in col]
# Extend to features list
features.extend(neighborhood_cols)
```

#### Assign `X`, feature matrix, and `y`, target vector, `saleprice`, for linear model.

```python
X = train[features]
y = train['saleprice_log']
```

### Baseline Model
- Create `y_mean` column to compare our model performance to the baseline model's.
- The baseline model takes the mean of the target vector and sets that as the prediction.

```python
# Baseline model
train['saleprice_mean'] = train['saleprice'].mean()
```

Find the $R^2$ score of baseline model.

```python
r2_score(train['saleprice'], train['saleprice_mean'])
```

> $R^2$ score of 0% indicates that the model explains none of the variability of the data.

#### Graph Predicted Values vs. Actual Values for Baseline

```python
# Make a scatter plot for predicted baseline values vs actual values
sns.lmplot(x='saleprice_mean', y='saleprice', data=train, height=5, aspect=2)
# Title and axis labels.
plt.xlabel("Predicted Values (Baseline): $\hat{y}$", fontsize = 20)
plt.ylabel("Actual Values: $y$", fontsize = 20)
plt.title('Predicted Values vs. Actual Values', fontsize = 24);
plt.savefig('../images/baseline.png');
```

> This `sns.lmplot()` shows the line of best fit is a horizontal line, meaning the baseline model does not explain the variability in the data.

___
### Linear Regression Model
- Instantiate model
- Fit the model
- Train, test, split

```python
# Istantiate model
lr = LinearRegression()

# train, test, split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# fit model
lr.fit(X_train, y_train)
```

Find **coefficients**
- Since I am using the log of the target vector, must exponentiat coefficients

```python
np.exp(lr.coef_)
```

Training $R^2$ score.

```python
lr.score(X_train, y_train)
```

> 85% of the variability in my data can be explained by the model

Testing $R^2$ score.

```python
lr.score(X_test, y_test)
```

> 85% of the variability in my data can be explained by the model

Cross validation score mean.

```python
cross_val_score(lr, X_train, y_train, cv=5).mean()
```

Predict vector target based on `X_train`.

```python
y_preds = np.exp(lr.predict(X_train))
```

Find **RMSE**.

```python
from sklearn import metrics

rmse = np.sqrt(metrics.mean_squared_error(np.exp(y_train), y_preds))

rmse
```

>Based on my Linear Reg, the model can predict the house's price is within $31,987.88 on avereage.

### Graphing the Linear Regression Coefficients.

```python
sns.set(style='whitegrid', font_scale = 1.5, rc={"figure.figsize": (15, 15)})
sns.barplot(x=lr.coef_, y=X_train.columns, palette='husl');
plt.title('Linear Regression Coefficient by Model Feature', fontsize=30)
plt.xlabel('Coefficient', fontsize=26)
plt.savefig('../images/linear_coef.png');
```

### Graphing Predicted Values vs. Actual Values for Linear Model

```python
# Code from 3.01-lesson-linear_regression 
# Plot the model
plt.figure(figsize=(12,9))

# Generate a scatterplot of predicted values versus actual values.
plt.scatter(y_preds, np.exp(y_train), s=5, color='skyblue', alpha = 0.8)
# Plot a line.

plt.plot([0, np.exp(np.max(y))],
         [0, np.exp(np.max(y))],
         color = 'black')

# Tweak title and axis labels.
plt.xlabel("Predicted Values: $\hat{y}$" + f'\n\n $R^2$ Score: {lr.score(X_test, y_test)}', fontsize = 20)
plt.ylabel("Actual Values: $y$", fontsize = 20)
plt.title('Linear Model Predicted Values vs. Actual Values', fontsize = 24);
plt.savefig('../images/linear.png');
```

### Linear Regression for Kaggle Data
- Submission for linear regression

### Standardize

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
```

```python
# Instantiate
sc = StandardScaler()

# It's learning the means and stds of each of our features
sc.fit(X_train)

Z_train = sc.fit_transform(X_train) # turned X_train into z-scores

# It's fine to transform X_test
Z_test = sc.transform(X_test) # turned X_test into z-scores
```

### Ridge for my model

```python
from sklearn.linear_model import RidgeCV
```

```python
# Set up a list of ridge alphas to check.
# Generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.
alphas_ridge = np.logspace(0, 5, 100)

# Cross-validate over our list of ridge alphas.
ridge_cv = RidgeCV(
    alphas= alphas_ridge,
    scoring='r2',
    cv=5
)

# Fit model using best ridge alpha!
ridge_cv.fit(Z_train, y_train)
```

```python
# Here is the optimal value of alpha
ridge_cv.alpha_
```

Our `ridge_model` object is actually already the model witht he optimal $\alpha$. Let's get the corresponding value of $R^2$.

```python
ridge_cv.score(Z_train, y_train)
```

>85% of the variability in my data can be explained by the model

```python
ridge_cv.score(Z_test, y_test)
```

>85% of the variability in my data can be explained by the model

### Graphing Predicted Values vs. Actual Values for Ridge Model

```python
y_preds_ridge = np.exp(ridge_cv.predict(Z_train))
```

```python
# Code from 3.01-lesson-linear_regression 
# Plot the model
plt.figure(figsize=(12,9))

# Generate a scatterplot of predicted values versus actual values.
plt.scatter(y_preds_ridge, np.exp(y_train), s=5, color='skyblue', alpha = 0.8)
# Plot a line.

plt.plot([0, np.exp(np.max(y))],
         [0, np.exp(np.max(y))],
         color = 'black')

# Tweak title and axis labels.
plt.xlabel("Predicted Values: $\hat{y}$" + f'\n\n $R^2$ Score: {ridge_cv.score(Z_test, y_test)}', fontsize = 20)
plt.ylabel("Actual Values: $y$", fontsize = 20)
plt.title('Ridge Model Predicted Values vs. Actual Values', fontsize = 24);
plt.savefig('../images/ridge.png');
```

### Lasso for my model

```python
from sklearn.linear_model import LassoCV
```

```python
# Set up a list of Lasso alphas to check.
alphas_lasso = np.logspace(-3, 0, 100)

# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV(
    alphas= alphas_lasso,
    cv=5)

# Fit model using best ridge alpha!
lasso_cv.fit(Z_train, y_train)
```

```python
# Here is the optimal value of alpha
lasso_cv.alpha_
```

```python
lasso_cv.score(Z_train, y_train)
```

>85% of the variability in my data can be explained by the model

```python
lasso_cv.score(Z_test, y_test)
```
